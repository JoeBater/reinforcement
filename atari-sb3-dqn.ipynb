```python
# train a model to play the game
import gymnasium as gym
print(gym.__version__)

from stable_baselines3 import PPO, DQN, A2C

env = gym.make("ALE/ChopperCommand-v5")
env = gym.wrappers.ResizeObservation(env, (125, 125)) # achieve an nxn square for the observation space
env = gym.wrappers.GrayScaleObservation(env, keep_dim = True) # convert to grayscale keeping the "color channel" dimension for now

env.unwrapped.seed(123)
# attempting 5M steps results in a memory error
total_timesteps = 1_000_000

# train the model using PPO using a policy gradient method
model = DQN('MlpPolicy', env, learning_rate=1e-3, verbose=1)
model.learn(total_timesteps)  # will train model over total_timesteps and report episode parameters 
model.save("ChopperCommand_125x125_gray_DQN_Mlp_1M")

# play the game using the trained model, vec_env is used to run multiple environments in parallel
vec_env = model.get_env()
obs = vec_env.reset()

terminated = False
rewards = 0
while terminated == False:
    # get the next action given the current observation
    action, _states = model.predict(obs, deterministic=False)
    obs, reward, terminated, info = model.env.step(action)  # model.env is the environment used by the model
    rewards += reward
    #print(rewards)

env.close()
print(rewards)
```

```python
# using stable-baselines3
import gymnasium as gym
from stable_baselines3 import PPO, DQN, A2C

env = gym.make("ALE/ChopperCommand-v5", render_mode="human")

env = gym.wrappers.ResizeObservation(env, (125, 125)) # achieve an nxn square for the observation space
env = gym.wrappers.GrayScaleObservation(env, keep_dim = True) # convert to grayscale keeping the "color channel" dimension for now

model = DQN.load("ChopperCommand_125x125_gray_DQN_Mlp_1M", env=env)
vec_env = model.get_env()

obs = vec_env.reset()
terminated = False
rewards = 0
while terminated == False:
    action, _states = model.predict(obs, deterministic=False)
    obs, reward, terminated, info = model.env.step(action)
    rewards += reward

env.close()
```
