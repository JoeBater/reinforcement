```python
import gymnasium as gym

env = gym.make("ALE/Seaquest-v5", render_mode="human")
observation, info = env.reset(seed=42)

for _ in range(1000):
    action = env.action_space.sample()
    observation, reward, terminated, truncated, info = env.step(action)

    if terminated or truncated:
        observation, info = env.reset()

env.close()
```

```python
# using stable-baselines3
import gymnasium as gym
from stable_baselines3 import PPO, DQN, A2C

env = gym.make("ALE/Asteroids-v5", render_mode="human")

env = gym.wrappers.ResizeObservation(env, (125, 125)) # achieve an nxn square for the observation space
env = gym.wrappers.GrayScaleObservation(env, keep_dim = True) # convert to grayscale keeping the "color channel" dimension for now

model = PPO.load("Asteroids_125x125_gray_PPO_Mlp_1M", env=env)
vec_env = model.get_env()

obs = vec_env.reset()
terminated = False
rewards = 0
while terminated == False:
    action, _states = model.predict(obs, deterministic=False)
    obs, reward, terminated, info = model.env.step(action)
    rewards += reward
    print(rewards)

env.close()
```

```python
import gymnasium as gym
from stable_baselines3 import PPO, DQN, A2C

env = gym.make("ALE/Seaquest-v5")
env = gym.wrappers.ResizeObservation(env, (125, 125)) # achieve an nxn square for the observation space
env = gym.wrappers.GrayScaleObservation(env, keep_dim = True) # convert to grayscale keeping the "color channel" dimension for now

env.seed(123)
total_timesteps = 1000000

model = PPO("MlpPolicy", env, verbose = 1)
model.learn(total_timesteps)  # will train model over total_timesteps and report episode parameters 
model.save("Seaquest_125x125_gray_PPO_Mlp_1M")

vec_env = model.get_env()
obs = vec_env.reset()
terminated = False
rewards = 0
while terminated == False:
    action, _states = model.predict(obs, deterministic=False)
    obs, reward, terminated, info = model.env.step(action)
    rewards += reward
    print(rewards)

env.close()
```
